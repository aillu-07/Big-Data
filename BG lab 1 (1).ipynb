```python
import requests
from bs4 import BeautifulSoup

url="https://www.mccormick.northwestern.edu/computer-science/people/faculty"

response=requests.get(url)
response
```

```python
Soup=BeautifulSoup(response.text,'html.parser')
```

```python
professor_elements = Soup.select('.faculty-info')
professor_elements
```

```python
professor_names = []
research_interests = []
profile_urls = []
education_details = []



faculty_profiles=Soup.find_all("div",class_='faculty-info')

for profile in faculty_profiles:
    
    names=profile.find("h3")
    if names:
        name = names.text.strip()
        professor_names.append(name)

for name in professor_names:
    print(name)
```

```python
research_interests = []

faculty_profiles = Soup.find_all("div", class_="faculty-info")

for profile in faculty_profiles:
    Interests = profile.find("div", class_='modal-content')
    
    if Interests:
        research = Interests.find("h2").text.strip()
    else:
        research = "N/A"
    research_interests.append(research)

for research in research_interests:
    print(research)
```

```python
profile_urls=[]

faculty_profiles = Soup.find_all("div", class_="faculty-info")
    
for profile in faculty_profiles:
    profile_url = profile.find("a", href=True)["href"]
    profile_urls.append(profile_url)
    
    print(profile_urls)
```

```python
education_details = []

for profile in faculty_profiles:
    
    education = profile.find("div", class_="")
    if education:
        edu = education.find("h2").text.strip("a","href")
        education_details.append(edu)
    else:
        education_details.append("N/A")
    print(education_details) 
            
```
